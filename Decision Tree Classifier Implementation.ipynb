```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris
```

```python
iris.data
```

```python
iris.target
```

```python
# We can import iris dataset in from another source
import seaborn as sns
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
# independent and dependent features
X = df.iloc[:,:-1]
y = iris.target
```

```python
X
```

```python
y
```

```python
### Train Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42) 
```

```python
X_train
```

# POST PRUNNING

```python
from sklearn.tree import DecisionTreeClassifier

### Post Prunning
prun_tree_model = DecisionTreeClassifier() 
```

```python
prun_tree_model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(prun_tree_model,filled=True)
```

```python
### Remove Post Prunning
tree_model = DecisionTreeClassifier(max_depth=2)
tree_model.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(tree_model,filled=True)
```

```python
y_predict = tree_model.predict(X_test)
```

```python
y_predict
```

```python
from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_predict, y_test)
```

```python
score
```

```python
print(classification_report(y_predict,y_test))
```

## Observations: 
1. We cannot work with assumption to remove Post Prunning
2. For that we use Pre Prunning using Hyper parameter tuning

# PRE PRUNNING

```python
parameters = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter' : ['best', 'random'],
    'max_depth' : [1,2,3,4,5],
    'max_features' : ['auto', 'sqrt', 'log2']
} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
treemodel = DecisionTreeClassifier()
cv = GridSearchCV(treemodel,parameters,cv=5,scoring='accuracy')
```

```python
cv.fit(X_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred = cv.predict(X_test)
```

```python
score = accuracy_score(y_pred,y_test)
score
```

```python
print(classification_report(y_pred,y_test))
```
